```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


time_green = ctrl.Antecedent(np.arange(10, 51, 1), 'время зеленого света')
street_north_south = ctrl.Antecedent(np.arange(0, 91, 1), 'улица Север-Юг')
street_west_east = ctrl.Antecedent(np.arange(0, 91, 1), 'улица Запад-Восток')
time_change = ctrl.Consequent(np.arange(-20, 21, 1), 'время зел.света на улице СЮ')

time_green['1'] = fuzz.trapmf(time_green.universe, [10, 10, 20, 25])
time_green['2'] = fuzz.trapmf(time_green.universe, [20, 25, 35, 40])
time_green['3'] = fuzz.trapmf(time_green.universe, [35, 40, 50, 50])

street_north_south['1'] = fuzz.trapmf(street_north_south.universe, [0, 0, 12, 18])
street_north_south['2'] = fuzz.trapmf(street_north_south.universe, [15, 21, 31, 37])
street_north_south['3'] = fuzz.trapmf(street_north_south.universe, [34, 40, 50, 56])
street_north_south['4'] = fuzz.trapmf(street_north_south.universe, [53, 59, 69, 75])
street_north_south['5'] = fuzz.trapmf(street_north_south.universe, [72, 78, 90, 90])

street_west_east['1'] = fuzz.trapmf(street_west_east.universe, [0, 0, 12, 18])
street_west_east['2'] = fuzz.trapmf(street_west_east.universe, [15, 21, 31, 37])
street_west_east['3'] = fuzz.trapmf(street_west_east.universe, [34, 40, 50, 56])
street_west_east['4'] = fuzz.trapmf(street_west_east.universe, [53, 59, 69, 75])
street_west_east['5'] = fuzz.trapmf(street_west_east.universe, [72, 78, 90, 90])

time_change['1'] = fuzz.gaussmf(time_change.universe, 20, 7)
time_change['2'] = fuzz.gaussmf(time_change.universe, 0, 7)
time_change['3'] = fuzz.gaussmf(time_change.universe, -20, 7)

svetofor_ctrl = ctrl.ControlSystem([
    ctrl.Rule(time_green['1'] & street_north_south['1'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['1'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['1'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['1'] & street_west_east['4'], time_change['2']),
    ctrl.Rule(time_green['1'] & street_north_south['1'] & street_west_east['5'], time_change['2']),

    ctrl.Rule(time_green['1'] & street_north_south['2'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['2'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['2'] & street_west_east['4'], time_change['2']),
    ctrl.Rule(time_green['1'] & street_north_south['2'] & street_west_east['5'], time_change['2']),

    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['4'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['3'] & street_west_east['5'], time_change['2']),

    ctrl.Rule(time_green['1'] & street_north_south['4'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['4'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['4'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['4'] & street_west_east['4'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['4'] & street_west_east['5'], time_change['1']),

    ctrl.Rule(time_green['1'] & street_north_south['5'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['5'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['5'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['5'] & street_west_east['4'], time_change['1']),
    ctrl.Rule(time_green['1'] & street_north_south['5'] & street_west_east['5'], time_change['1']),

    ctrl.Rule(time_green['2'] & street_north_south['1'] & street_west_east['1'], time_change['2']),
    ctrl.Rule(time_green['2'] & street_north_south['1'] & street_west_east['2'], time_change['2']),
    ctrl.Rule(time_green['2'] & street_north_south['1'] & street_west_east['3'], time_change['3']),
    ctrl.Rule(time_green['2'] & street_north_south['1'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['2'] & street_north_south['1'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['2'] & street_north_south['2'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['2'] & street_west_east['2'], time_change['2']),
    ctrl.Rule(time_green['2'] & street_north_south['2'] & street_west_east['3'], time_change['3']),
    ctrl.Rule(time_green['2'] & street_north_south['2'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['2'] & street_north_south['2'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['2'] & street_north_south['3'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['3'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['3'] & street_west_east['3'], time_change['2']),
    ctrl.Rule(time_green['2'] & street_north_south['3'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['2'] & street_north_south['3'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['2'] & street_north_south['4'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['4'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['4'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['4'] & street_west_east['4'], time_change['2']),
    ctrl.Rule(time_green['2'] & street_north_south['4'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['2'] & street_north_south['5'] & street_west_east['1'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['5'] & street_west_east['2'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['5'] & street_west_east['3'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['5'] & street_west_east['4'], time_change['1']),
    ctrl.Rule(time_green['2'] & street_north_south['5'] & street_west_east['5'], time_change['2']),

    ctrl.Rule(time_green['3'] & street_north_south['1'] & street_west_east['1'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['1'] & street_west_east['2'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['1'] & street_west_east['3'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['1'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['1'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['3'] & street_north_south['2'] & street_west_east['1'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['2'] & street_west_east['2'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['2'] & street_west_east['3'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['2'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['2'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['3'] & street_north_south['3'] & street_west_east['1'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['3'] & street_west_east['2'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['3'] & street_west_east['3'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['3'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['3'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['3'] & street_north_south['4'] & street_west_east['1'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['4'] & street_west_east['2'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['4'] & street_west_east['3'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['4'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['4'] & street_west_east['5'], time_change['3']),

    ctrl.Rule(time_green['3'] & street_north_south['5'] & street_west_east['1'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['5'] & street_west_east['2'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['5'] & street_west_east['3'], time_change['2']),
    ctrl.Rule(time_green['3'] & street_north_south['5'] & street_west_east['4'], time_change['3']),
    ctrl.Rule(time_green['3'] & street_north_south['5'] & street_west_east['5'], time_change['3'])
])

svetofor_simulation = ctrl.ControlSystemSimulation(svetofor_ctrl)

svetofor_simulation.input['время зеленого света'] = 10
svetofor_simulation.input['улица Север-Юг'] = 10
svetofor_simulation.input['улица Запад-Восток'] = 15

svetofor_simulation.compute()

print("Изменение времени зеленого света на улице Север-Юг:", svetofor_simulation.output['время зел.света на улице СЮ'])
```

```python
import matplotlib.pyplot as plt

input_values = np.arange(1, 50, 1)
output_values = []

for input_value in input_values:
    svetofor_simulation = ctrl.ControlSystemSimulation(svetofor_ctrl)
    svetofor_simulation.input['время зеленого света'] = input_value
    svetofor_simulation.input['улица Север-Юг'] = 10
    svetofor_simulation.input['улица Запад-Восток'] = 10
    svetofor_simulation.compute()
    output_values.append(svetofor_simulation.output['время зел.света на улице СЮ'])

plt.plot(input_values, output_values)
plt.title('Output vs. Input')
plt.xlabel('Input (время зеленого света)')
plt.ylabel('Output (время зел.света на улице СЮ)')
plt.grid(True)
plt.show()
```
